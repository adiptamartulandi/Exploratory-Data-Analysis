### Mengisi missing values pada data

```python
count_missing_new = data.isnull().sum()
count_missing_new #menghitung value data baru yang null
```

```python
percentage_new = round((count_missing_new/len(data))*100,2)
percentage_new # persentase data yang missing pada tiap kolom
```

Cara mengisi missing values :
1. untuk kolom year, karena ini merupakan data numerical, maka diisi dengan median dari kolom year
2. untuk kolom manufacturer, karena ini merupakan data kategorik, maka diisi dengan mode dari kolom manufacturer
3. untuk kolom make, karena ini merupakan data kategorik, maka diisi dengan mode dari kolom make
4. untuk kolom fuel, karena ini merupakan data kategorik, maka diisi dengan mode dari kolom fuel
5. untuk kolom title_status, karena ini merupakan data kategorik, maka diisi dengan mode dari kolom title_status
6. untuk kolom transmission, karena ini merupakan data kategorik, maka diisi dengan mode dari kolom transmission
7. untuk kolom county_name, karena ini merupakan data kategorik, maka diisi dengan mode dari kolom county_name
8. untuk kolom weather, karena ini merupakan data numerical, maka diisi dengan median dari kolom weather

saya mengisi kolom yang mode dengan looping

```python
fill_mode = ['manufacturer', 'make', 'fuel', 'title_status', 'transmission', 'county_name']
for feature in fill_mode:
    if feature in data.columns.values:
        data[feature] = data[feature].fillna(data[feature].value_counts().index[0])
```

```python
fill_median = ['year', 'weather']
for feature in fill_median:
    if feature in data.columns.values:
        data[feature] = data[feature].fillna(data[feature].median())
```

```python
data2.corr(method='pearson').style.background_gradient().set_precision(2) #untuk semua data yang sudah di label encoder
```
